**MENGHITUNG PERSENTASE DATA MISSING**

```python
import pandas as pd

# Membaca dataset dari file CSV
df = pd.read_csv("dataset.csv", delimiter=";")


# Menghitung jumlah data yang hilang dalam setiap kolom
missing_data = df.isnull().sum()

# Menghitung jumlah total baris dalam dataset
total_data = len(df)

# Menghitung persentase data yang hilang dalam setiap kolom
percentage_missing = (missing_data / total_data) * 100

# Menampilkan persentase data yang hilang dalam setiap kolom
print(percentage_missing)
```

Persentase data missing setiap kolom rata rata di bawah 10% hal ini berarti data missing tidak terlalu signifikan maka langkah yang diambil untuk menghandel data missing ini adalah dengan menggunakan metode fillna untuk mengisi data data kosong tersebut.

**MENGHANDEL DATA MISSING**

```python
import pandas as pd

# Membaca dataset dari file CSV
df = pd.read_csv("dataset.csv", delimiter=";")

# Daftar kolom numerik
numeric_columns = ['Tenure', 'WarehouseToHome', 'HourSpendOnApp', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder']

# Imputasi missing value dalam kolom numerik dengan mean
for column in numeric_columns:
    df[column].fillna(df[column].mean(), inplace=True)
```

imputasi (pengisian) nilai yang hilang dalam kolom numerik menggunakan rata-rata (mean) karena data missing yang terdapat dalam dataset ini dalam bentu numeric. Maka penjelasannya dari setiap kolomnya adalah :

1. Tenure: Ini adalah kolom numerik yang mewakili lamanya pelanggan tetap menggunakan layanan. Dalam kode ini, jika ada nilai yang hilang dalam kolom "Tenure," nilai tersebut akan diisi dengan rata-rata (mean) dari seluruh data dalam kolom "Tenure."

2. WarehouseToHome: Ini juga adalah kolom numerik yang mewakili jarak antara gudang dan rumah pelanggan. Jika ada nilai yang hilang dalam kolom "WarehouseToHome," nilai tersebut akan diisi dengan rata-rata (mean) dari seluruh data dalam kolom "WarehouseToHome."

3. HourSpendOnApp: Ini adalah kolom numerik yang mewakili berapa banyak waktu yang dihabiskan pelanggan di aplikasi Anda. Jika ada nilai yang hilang dalam kolom "HourSpendOnApp," nilai tersebut akan diisi dengan rata-rata (mean) dari seluruh data dalam kolom "HourSpendOnApp."

4. OrderAmountHikeFromlastYear: Ini adalah kolom numerik yang mewakili kenaikan jumlah pesanan dari tahun sebelumnya. Jika ada nilai yang hilang dalam kolom ini, nilai tersebut akan diisi dengan rata-rata (mean) dari seluruh data dalam kolom ini.

5. CouponUsed: Ini adalah kolom numerik yang mewakili jumlah kupon yang digunakan pelanggan. Jika ada nilai yang hilang dalam kolom ini, nilai tersebut akan diisi dengan rata-rata (mean) dari seluruh data dalam kolom ini.

6. OrderCount: Ini adalah kolom numerik yang mewakili jumlah pesanan yang dibuat oleh pelanggan. Jika ada nilai yang hilang dalam kolom ini, nilai tersebut akan diisi dengan rata-rata (mean) dari seluruh data dalam kolom ini.

7. DaySinceLastOrder: Ini adalah kolom numerik yang mewakili berapa hari sejak pesanan terakhir pelanggan. Jika ada nilai yang hilang dalam kolom ini, nilai tersebut akan diisi dengan rata-rata (mean) dari seluruh data dalam kolom ini.

# 2B Feature extraction (membuat feature baru dari feature yang sudah ada)

```python
#import library yang dibutuhkan
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

### Feature baru 1

```python
#membuat feature tambahan baru 1
df['TotalSpending'] = df['OrderAmountHikeFromlastYear'] * df['OrderCount']
TotalSpending = df['TotalSpending'].to_frame()
TotalSpending
```

TotalSpending : hasil penggabungan fitur "OrderAmountHikeFromlastYear" dan "OrderCount" untuk membuat fitur yang mengukur total pengeluaran pelanggan dalam setahun. Feature ini dapat memberikan wawasan tentang tingkat pengeluaran pelanggan.

### Feature baru 2

```python
#membuat feature tambahan baru 2
df['AverageOrderValue'] = df['OrderAmountHikeFromlastYear'] / df['OrderCount']
AverageOrderValue = df['AverageOrderValue'].to_frame()
AverageOrderValue
```

AverageOrderValue : hasil perhitungan rata-rata nilai pesanan dengan membagi "OrderAmountHikeFromlastYear" dengan "OrderCount". Feature ini ni dapat memberikan informasi tentang seberapa besar setiap pesanan yang ditempatkan oleh pelanggan.

### Feature baru 3

```python
#membuat feature tambahan baru 3
df['CashbackPercentage'] = (df['CashbackAmount'] / df['OrderAmountHikeFromlastYear']) * 100
CashbackPercentage = df['CashbackPercentage'].to_frame()
CashbackPercentage
```

CashbackPercentage: hasil perhitungan persentase uang kembalian berdasarkan "CashbackAmount" dibandingkan dengan total pengeluaran ("OrderAmountHikeFromlastYear").

### Feature baru 4

```python
#membuat feature tambahan baru 4
df['FrequencyOfOrders'] = df['OrderCount'] / df['Tenure']
FrequencyOfOrders = df['FrequencyOfOrders'].to_frame()
FrequencyOfOrders
```

FrequencyOfOrders: fitur yang mengukur seberapa sering pelanggan melakukan pesanan dengan membagi "OrderCount" dengan "Tenure." Feature ini berguna untuk memberitahu tentang kebiasaan pemesanan pelanggan.
