# B Feature extraction (membuat feature baru dari feature yang sudah ada)

### Feature baru 1

```python
#membuat feature tambahan baru 1
df['TotalSpending'] = df['OrderAmountHikeFromlastYear'] * df['OrderCount']
TotalSpending = df['TotalSpending'].to_frame()
TotalSpending
```

TotalSpending : hasil penggabungan fitur "OrderAmountHikeFromlastYear" dan "OrderCount" untuk membuat fitur yang mengukur total pengeluaran pelanggan dalam setahun. Feature ini dapat memberikan wawasan tentang tingkat pengeluaran pelanggan.

### Feature baru 2

```python
#membuat feature tambahan baru 2
df['AverageOrderValue'] = df['OrderAmountHikeFromlastYear'] / df['OrderCount']
AverageOrderValue = df['AverageOrderValue'].to_frame()
AverageOrderValue
```

AverageOrderValue : hasil perhitungan rata-rata nilai pesanan dengan membagi "OrderAmountHikeFromlastYear" dengan "OrderCount". Feature ini ni dapat memberikan informasi tentang seberapa besar setiap pesanan yang ditempatkan oleh pelanggan.

### Feature baru 3

```python
#membuat feature tambahan baru 3
df['CashbackPercentage'] = (df['CashbackAmount'] / df['OrderAmountHikeFromlastYear']) * 100
CashbackPercentage = df['CashbackPercentage'].to_frame()
CashbackPercentage
```

CashbackPercentage: hasil perhitungan persentase uang kembalian berdasarkan "CashbackAmount" dibandingkan dengan total pengeluaran ("OrderAmountHikeFromlastYear").

### Feature baru 4

```python
#membuat feature tambahan baru 4
df['FrequencyOfOrders'] = df['OrderCount'] / df['Tenure']
FrequencyOfOrders = df['FrequencyOfOrders'].to_frame()
FrequencyOfOrders
```

```python
#membuat feature tambahan baru 5
df['DeviceUtilization'] = df['HourSpendOnApp'] / df['NumberOfDeviceRegistered']
DeviceUtilization = df['DeviceUtilization'].to_frame()
DeviceUtilization
```

```python
#membuat feature tambahan baru 6
df['CouponUsageRate'] = df['CouponUsed'] * 100 / df['OrderCount']
CouponUsageRate = df['CouponUsageRate'].to_frame()
CouponUsageRate
```

FrequencyOfOrders: fitur yang mengukur seberapa sering pelanggan melakukan pesanan dengan membagi "OrderCount" dengan "Tenure." Feature ini berguna untuk memberitahu tentang kebiasaan pemesanan pelanggan.

```python
#menggabungkan semua feature baru menjadi 1 dataframe
concatenated_df = pd.concat([TotalSpending, AverageOrderValue, FrequencyOfOrders, CashbackPercentage, DeviceUtilization, CouponUsageRate], axis=1)
concatenated_df
```

```python
#mengambil dataframe target 'Churn'
concatenated_df['Churn'] = df['Churn']
churn = concatenated_df['Churn'].to_frame()
churn
```

```python
#menggabungkan dataframe feature-feature baru dengan dataframe target/'Churn'
concatenated_df_churn = pd.concat([churn, TotalSpending, AverageOrderValue, FrequencyOfOrders, CashbackPercentage], axis=1)
concatenated_df_churn
```

Tindakan yang dilakukan:
- Setelah membandingkan kolerasi antara feature 'TotalSpending' dan 'OrderCount' dengan 'Churn'. Maka feature 'TotalSpending' akan dihapus/tidak jadi digunakan karena memiliki kolerasi yang sama yakni -0.03 (lebih baik menggunakan feature yang sudah ada dengan nilai kolerasi yang sama)
- Setelah membandingkan kolerasi antara feature 'CashbackPercentage' dan 'CashbackAmount' dengan 'Churn'. Maka feature 'CashbackPercentage' akan dihapus/tidak jadi digunakan karena memiliki kolerasi yang rendah dengan 'Churn'.
